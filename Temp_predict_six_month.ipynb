```python
import matplotlib.pyplot as plt

# Dữ liệu đã cho: Nhiệt độ cao trung bình tại Kansas
Nhiệt_độ_cao = [37, 44, 55, 66, 75, 84, 89, 88, 80, 69, 53, 41]

# Vẽ đồ thị các nhiệt độ
tháng = list(range(1, 13))
plt.plot(tháng, Nhiệt_độ_cao,'g-x') #Ta có g-x có nghĩa là green , - chỉ đánh dấu điểm trên biểu đồ trong trường hợp này điểm trên biểu đồ được kí hiệu là chữ X
plt.title('Nhiệt độ cao trung bình tại Kansas', )
plt.xlabel('Tháng')
plt.ylabel('Nhiệt độ (F)')
plt.axis([1, 12, 30, 100])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

WeatherHigh = np.array([37, 44, 55, 66, 75, 84, 89, 88, 80, 69, 53, 41])
plt.figure()
plt.plot(range(1, 13), WeatherHigh, 'r-x')
plt.axis([1, 12, 30, 100])

x = np.array([1, 3, 5, 8, 10, 12])  # Các tháng: January, May, August, và December
V = np.vander(x)
#Hàm X có độ dài là 4 => len(x)  = 4 trong khi ta muốn độ dài của matrix weatherHigh = 4 x4 thì phải trừ 1 còn 3 vì trong python tính từ 0 1 2 3 =) có độ dài = 4
print(V)
# Giải hệ phương trình tuyến tính Ax = b trong đó A = V , b= WeatherHigh


CoefHigh = np.linalg.solve(V, WeatherHigh[x-1]) # Này là lấy WeatherHigh của các tháng 1 5 8 12 => [37 75 88 41]
print(WeatherHigh[x-1])


print(CoefHigh)
```

```python
xc = np.arange(1, 12.1, 0.1)
print(xc)
ycHigh = np.polyval(CoefHigh, xc)  
plt.plot(xc, ycHigh, 'b-|')
plt.plot(np.arange(1, 13), WeatherHigh, 'r-x')
plt.axis([1, 12, 30, 95])
plt.show()
```
