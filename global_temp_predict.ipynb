```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, PchipInterpolator

filename = 'global_annual_mean_temp_anomalies_land-ocean_1880-2016.csv'
dataFrame = pd.read_csv(filename)

Year = dataFrame['Year'].values
No_Smoothing = dataFrame['No_Smoothing'].values
Lowess = dataFrame['Lowess'].values

xc = np.linspace(min(Year), max(Year), 100)


yc_NoSmoothing_linear = interp1d(Year, No_Smoothing, kind='linear')(xc) # th vin interp1d l th vin ni suy. 
yc_NoSmoothing_spline = interp1d(Year, No_Smoothing, kind='cubic')(xc)
pchip_interpolator = PchipInterpolator(Year, No_Smoothing)
yc_NoSmoothing_pchip = pchip_interpolator(xc)

yc_Lowess_linear = interp1d(Year, Lowess, kind='linear')(xc)
yc_Lowess_spline = interp1d(Year, Lowess, kind='cubic')(xc)
pchip_interpolator = PchipInterpolator(Year, Lowess)
yc_Lowess_pchip = pchip_interpolator(xc)

plt.figure(figsize=(10, 12))
plt.plot(Year, No_Smoothing, 'b-x', label='No Smoothing')
plt.plot(xc, yc_NoSmoothing_linear, label='Linear Interpolation')
plt.plot(xc, yc_NoSmoothing_pchip, label='PCHIP Interpolation')
plt.plot(xc, yc_NoSmoothing_spline, label='Spline Interpolation')
plt.title('Global Annual Mean Temperature Anomalies (No Smoothing)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (F)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filename = 'global_annual_mean_temp_anomalies_land-ocean_1880-2016.csv'
dataFrame = pd.read_csv(filename)

years = dataFrame['Year'].values
temp = np.array(dataFrame['No_Smoothing'])

B2 = np.column_stack((np.ones_like(years), years))

Q2 = np.linalg.qr(B2)[0]

print("Rank of Q2:", np.linalg.matrix_rank(Q2))
print("Rank of [Q2, B2]:", np.linalg.matrix_rank(np.column_stack((Q2, B2))))

P2 = Q2.dot(Q2.T)

temp2 = P2.dot(temp)

plt.figure(figsize=(10, 6))
plt.plot(years, temp, label='Original Data')
plt.plot(years, temp2, label='Linear Approximation', linestyle='--')
plt.xlabel('Year')
plt.ylabel('Temperature Anomalies')
plt.title('Global Annual Mean Temperature Anomalies (1880-2016)')
plt.legend()
plt.grid(True)
plt.axis([1880, 2020, -0.5, 1.5])
X = np.arange(min(Year), max(Year) + 20 , 20)
print(X)
Y = np.arange(-0.5, 1.5 , 0.5)
print(Y)
plt.xticks(X)
plt.yticks(Y)
plt.show()

print("Norm of (P2 * P2 - P2):", np.linalg.norm(P2.dot(P2) - P2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Hm tm h s ti u cho m hnh tuyn tnh
def find_optimal_coefficients_linear(years, temp):
    B2 = np.column_stack((np.ones_like(years), years)) # Phng trnh tuyn tnh bc 1 y = a + bx : y l temp, x l future year, phi vit l a.1 +bx th mi ng  
    # print("B2 \n", B2)x
    #P2 = Q2Q2T

    coefficients = np.linalg.lstsq(B2, temp, rcond=None)[0] #Gii h phng trnh tuyn tnh bng phng php bnh phng ti thiu. => pt y = a+bx ||  trong  y l temp ,a l year,  b l  1
    #lstqs => a v W = (A^T*A)^-1 A^T*y || B2 l A . temp = y 
    print("B2: " , coefficients)
    return coefficients

# Hm tm h s ti u cho m hnh bc hai
def find_optimal_coefficients_quadratic(years, temp):
    B3 = np.column_stack((np.ones_like(years), years, years**2)) # ( 1, x , x^2) => a + bx + cx^2
    # print("B3 \n", B3)
    coefficients = np.linalg.lstsq(B3, temp, rcond=None)[0] # y = a + bx + cx^2 :: x^2 = futureyear ^2 , ....   , tm a ,b , c
    print("B3: " , coefficients)
    return coefficients

# c d liu t file CSV
filename = 'global_annual_mean_temp_anomalies_land-ocean_1880-2016.csv'
dataFrame = pd.read_csv(filename)

years = dataFrame['Year'].values
temp = np.array(dataFrame['No_Smoothing'])

# Tm h s cho m hnh tuyn tnh v bc hai
coefficients_linear = find_optimal_coefficients_linear(years, temp)
coefficients_quadratic = find_optimal_coefficients_quadratic(years, temp)

# Tnh ton gi tr d on cho cc nm tng lai
future_years = np.arange(2016, 2026)
predicted_temp_linear = coefficients_linear[0] + coefficients_linear[1] * future_years
predicted_temp_quadratic = (coefficients_quadratic[0] +
                            coefficients_quadratic[1] * future_years +
                            coefficients_quadratic[2] * future_years**2)

# V biu 
plt.figure(figsize=(10, 6))
plt.plot(years, temp, 'o', label='Temperature Data')
plt.plot(years, (coefficients_quadratic[0] +
                 coefficients_quadratic[1] * years +
                 coefficients_quadratic[2] * years**2), '-', label='Quadratic Approximation')
plt.plot(future_years, predicted_temp_quadratic, 'g--', label='Predicted Temperature Anomalies (Quadratic)')
plt.plot(years, coefficients_linear[0] + coefficients_linear[1] * years, '-', label='Linear Approximation')
plt.plot(future_years, predicted_temp_linear, 'r--', label='Predicted Temperature Anomalies (Linear)')

plt.xlabel('Year')
plt.ylabel('Temperature Anomalies')
plt.title('Global Annual Mean Temperature Anomalies (1880-2016)')
plt.legend()
plt.grid(True)
plt.axis([1880, 2025, -0.5, 1.5])
plt.xticks(np.arange(1880, 2026, 20))
plt.yticks(np.arange(-0.5, 1.6, 0.5))
plt.show()
```
